```python
df.groupby(df.school).count()
```

```python
school_results = df.groupby(by=[df.school, df.passed]).size().reset_index(name='count')
school_results
```

```python
school_results['percentage'] = school_results.groupby('school')['count'].transform(lambda x: (x / x.sum()) * 100)
```
